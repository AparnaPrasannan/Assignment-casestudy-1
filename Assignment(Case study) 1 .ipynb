```python
## 5.Visualizations
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
## 1. Histogram
```

```python
plt.figure(figsize=(10,7))
plt.hist(ds['SL'])
plt.title("SL map",c='green',fontsize='20')
plt.xlabel('Value')
plt.ylabel('Numbers')
plt.show()
```

```python
## 2.Countplot
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=ds['SL'])
plt.title('Countplot on SL')
plt.show()
```

```python
## 3. Violin plot
```

```python
plt.figure(figsize=(26,8))
sns.violinplot(x=ds['SL'],y=ds['SW'],z=ds['PL'],hue=ds['Classification'])
plt.title('Violin plot',fontsize='25')
plt.xlabel('SL',fontsize='15')
plt.ylabel('SW',fontsize='15')
plt.show()
```
